```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import time
import os
import dask.dataframe as dd
```

```python
#Process the large dataset in chunks to create a cleaned version

chunk_size = 200000  # Number of rows per chunk
cleaned_filename = 'cleaned_transactions_dataset.csv'

print(f"Starting to process the dataset in chunks of {chunk_size} rows...")
start_time = time.time()

# Remove the cleaned file if it exists
if os.path.exists(cleaned_filename):
    os.remove(cleaned_filename)

# Set up a counter to track progress
chunk_counter = 0

# Create an iterator that reads the CSV in chunks
for chunk in pd.read_csv('large_transactions_dataset.csv', chunksize=chunk_size):
   
    # Handle missing transaction_amount values by filling with the median of this chunk
    median_amount = chunk['transaction_amount'].median()
    chunk['transaction_amount'].fillna(median_amount, inplace=True)
    
    # Determine if this is the first chunk (to write headers)
    is_first_chunk = (chunk_counter == 0)
    
    # Append the cleaned chunk to the new file
    chunk.to_csv(cleaned_filename, mode='a', header=is_first_chunk, index=False)
    
    # Progress indicator
    chunk_counter += 1
    if chunk_counter % 5 == 0:
        print(f"  ... {chunk_counter} chunks processed")

end_time = time.time()
cleaned_file_size = os.path.getsize(cleaned_filename) / (1024 * 1024)  # in MB

print("\n--- Cleaning Complete! ---")
print(f"Cleaned file '{cleaned_filename}' created successfully.")
print(f"Original file size: {os.path.getsize('large_transactions_dataset.csv') / (1024 * 1024):.2f} MB")
print(f"Cleaned file size: {cleaned_file_size:.2f} MB")
print(f"Total chunks processed: {chunk_counter}")
print(f"Time taken: {end_time - start_time:.2f} seconds")


```

```python
# Verify the cleaned dataset
print("Verifying the cleaned dataset...")

# Read a small sample to check the cleaning worked
sample_df = pd.read_csv('cleaned_transactions_dataset.csv', nrows=10000)

print(f"Sample of cleaned data:")
print(f"Shape: {sample_df.shape}")
print(f"Missing values in transaction_amount: {sample_df['transaction_amount'].isnull().sum()}")
print(f"Data types:")
print(sample_df.dtypes)
print(f"\nFirst few rows:")
print(sample_df.head())
```

```python
#Display the actual csv data
data = pd.read_csv('cleaned_transactions_dataset.csv')
print(data.head())
```

```python
#Ideas to Investigate:

#Category Performance: What is the average transaction_amount for each product_category? 
#Which category is the most valuable on average? (A bar chart is great for this).

#Average Transaction Amount by Category
avg_amount_by_category = data.groupby('product_category')['transaction_amount'].mean().sort_values(ascending=False)
sns.barplot(x=avg_amount_by_category.index, y=avg_amount_by_category.values, palette='viridis')
plt.title('Average Transaction Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Transaction Amount')
plt.xticks(rotation=45)
plt.show()


#Statistical Summary: Average transaction_amount by product_category
print("\nStatistical Summary of transaction_amount by product_category:")
print("\nAverage transaction amount by category")
print(avg_amount_by_category)

#Which product_category has the highest number of average transactions?
avg_transactions_by_category = data['product_category'].value_counts().sort_values(ascending=False)
max_category = avg_transactions_by_category.idxmax()
max_count = avg_transactions_by_category.max()
print(f"\nProduct category with the highest number of transactions: {max_category} ({max_count} transactions)")
```

```python
#Fraud Analysis: Which product_category has the highest number of fraudulent transactions? 
#(Use a countplot after filtering for fraudulent transactions).

# Filter for fraudulent transactions
fraudulent_df = data[data['is_fraudulent'] == 1]

print(f"Total transactions: {len(data):,}")
print(f"Fraudulent transactions: {len(fraudulent_df):,}")
print(f"Fraud rate: {len(fraudulent_df)/len(data)*100:.2f}%")

# Visualize fraudulent transactions by category
plt.figure(figsize=(12, 6))
sns.countplot(data=fraudulent_df, x='product_category', 
              order=fraudulent_df['product_category'].value_counts().index, 
              palette='Reds')
plt.title('Number of Fraudulent Transactions by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Fraudulent Transactions')
plt.xticks(rotation=45)
plt.show()

# Print statistics
fraud_by_category = fraudulent_df['product_category'].value_counts()
highest_fraud_category = fraud_by_category.index[0]
highest_fraud_count = fraud_by_category.iloc[0]


print("\nFraudulent transactions by category:")
print(fraud_by_category)

print(f"\nProduct category with the highest number of fraudulent transactions: {highest_fraud_category} ({highest_fraud_count} transactions)")
```

```python
#Time-Based Patterns: Create a new column for the month from the timestamp. 

data['transaction_month'] = pd.to_datetime(data['timestamp']).dt.month
monthly_transactions = data['transaction_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_transactions.index, y=monthly_transactions.values, palette='Blues')
plt.title('Number of Transactions by Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.show()

#How does the number of transactions change throughout the year?
print("\nMonthly transaction counts:")
print(monthly_transactions)

#Conclusion for how does the number of transactions change throughout the year?
print("\nThe number of transactions varies by month, with some months showing significantly higher activity than others.")
print("This could be due to seasonal trends, holidays, or promotional events that drive consumer spending.")
```
